```python
import numpy as np
from scipy.linalg import solve_continuous_are, eig
import matplotlib.pyplot as plt
```

```python
# we have to find X' as a function of X and t 
total_time = 10
dt = 0.01

itr = total_time/dt
x_list = []
print(itr)

pie = 3.14
x0 = np.array([2.14,1])
x = x0
B = np.array([[0],[2]])
Initial_A = np.array([[0,1],[10,-0.5]])
Q = np.diag([1,1])
R = np.array([[0.01]])
A = Initial_A
thetas = []
iterations = []
```

```python
for i in range(int(itr)):
    # we have A and B which can as taken as constants in that particular t , so that equation converts to X' = Ax + Bu
    P = solve_continuous_are(A, B, Q, R)
    K = np.linalg.inv(R) @ B.T @ P
    u = -K @ x
    xdot = A @ x + B @ u
    next_x =  x + xdot * dt
    A = np.array([[0,1],[(10*np.sin(x[0])),-0.5]])
    x = next_x
    thetas.append(pie-x[0])
    iterations.append(i+1)
    print(x)
```

```python
plt.plot(iterations,thetas)
plt.xlabel('Update Steps')
plt.ylabel('Theta value')
plt.title('SDRE')
plt.show()
```
